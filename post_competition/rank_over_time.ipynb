```python
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('public_ranks.csv', parse_dates='date', index_col='date')
# first two entries were trash
df = df.sort_index().iloc[2:]
```

```python
df.iloc[:10]
```

```python
dates = df.index.to_pydatetime()
scores = df['public_score'].as_matrix()

pl.figure(figsize=(12, 4))
pl.title('Public leaderboard score over time')
pl.plot(scores)
pl.xticks(np.arange(scores.shape[0])[::5], dates[::5], rotation=45);
pl.axvline(scores.argmax(), c='r')
```
